```python
class Car:
    def __init__(self,engine,windows=4,doors=4):
        self.windows = windows
        self.doors = doors
        self.engine = engine
        pass
    def drive(self):
        print(f"Driving {self.engine}")
```

```python
porche = Car("v8")
porche.drive()
```

```python
class Tesla(Car): # Tesla(Car,....) Multiple Inheritance
    def __init__(self, engine, windows=4, doors=4, is_self_driving=True):
        super().__init__(engine, windows, doors)
        self.is_self_driving = is_self_driving

    def toggle_self_driving(self):
        self.is_self_driving = not self.is_self_driving
        print(self.is_self_driving)

T1 = Tesla("Stututututtu")
T1.toggle_self_driving()
```
