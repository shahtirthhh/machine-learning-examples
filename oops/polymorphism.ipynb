```python
class Animal:
    def speak(self):
        return "Animal speaking..."

class Dog(Animal):
    def speak(self):
        return "Whooffff!!!"
class Cat(Animal):
    def speak(self):
        return "Meowwwww.."

k = Cat()
k.speak()
```

```python
# Interfaces -> Abstract base classes

from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    pass
        
```
